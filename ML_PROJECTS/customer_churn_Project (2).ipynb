3. Linear Regression:
● Build a simple linear model where dependent variable is ‘MonthlyCharges’
and independent variable is ‘tenure’:
a. Divide the dataset into train and test sets in 70:30 ratio.
b. Build the model on train set and predict the values on test set
c. After predicting the values, find the root mean square error
d. Find out the error in prediction & store the result in ‘error’
e. Find the root mean square error


```python
x=df[['tenure']]
y=df[['MonthlyCharges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=7)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss
```

```python
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
linmodel=LinearRegression()
linmodel
```

```python
linmodel.fit(x_train,y_train)
```

```python
y_pred=linmodel.predict(x_test)
y_pred
```

```python
y_pred=pd.DataFrame(y_pred,columns=['predict_values'])
y_pred
```

```python
y_test
```

```python
data=pd.DataFrame(columns=['Actual','Predict'])
data
```

```python
data['Actual']=y_test.reset_index()['MonthlyCharges']
data['Predict']=y_pred.reset_index()['predict_values']
```

```python
data
```

```python
data['error']=data['Actual']-data['Predict']
data
```

```python
from sklearn import metrics
print("root mean square error ",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
y_pred=rfc.predict(x_test)
y_pred
```
